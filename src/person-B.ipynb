```python
import pandas as pd

df = pd.read_csv("AmesHousing.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
important_vars = ["SalePrice", "Gr Liv Area", "Overall Qual", "Year Built"]
df[important_vars].describe()

df["SalePrice"].quantile([0.25, 0.5, 0.75])
```

```python
df = df.drop(["Order", "PID"], axis=1)
```

```python
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols = df.select_dtypes(include=["object"]).columns
```

```python
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

for col in cat_cols:
    df[col] = df[col].fillna("None")
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
X = df.drop("SalePrice", axis=1)
y = df["SalePrice"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lin_model = LinearRegression()
lin_model.fit(X_train, y_train)

y_pred_lin = lin_model.predict(X_test)

mse_lin = mean_squared_error(y_test, y_pred_lin)
rmse_lin = mse_lin ** 0.5
r2_lin = r2_score(y_test, y_pred_lin)

print("Linear RMSE:", rmse_lin)
print("Linear R2:", r2_lin)
```

```python
import statsmodels.api as sm
import pandas as pd

X_train_sm = sm.add_constant(X_train)

# Ensure all columns are numeric, coercing errors to NaN
# This will handle any 'object' dtype columns that contain non-numeric strings
# or mixed types by turning non-convertible entries into NaN.
for col in X_train_sm.columns:
    X_train_sm[col] = pd.to_numeric(X_train_sm[col], errors='coerce')

# Drop any columns that became entirely NaN after coercion (i.e., contained no numeric data)
X_train_sm = X_train_sm.dropna(axis=1, how='all')

# Fill any remaining NaNs (introduced by coercion for individual non-numeric values)
# For simplicity, filling with 0. A more sophisticated imputation could be used if needed.
X_train_sm = X_train_sm.fillna(0)

# Explicitly convert X_train_sm and y_train to NumPy arrays of float type
# This ensures that statsmodels receives purely numeric data.
X_train_sm_numeric = X_train_sm.values.astype(float)
y_train_numeric = y_train.values.astype(float)

ols_model = sm.OLS(y_train_numeric, X_train_sm_numeric).fit()

print("AIC:", ols_model.aic)
print("BIC:", ols_model.bic)
```

```python
from sklearn.linear_model import RidgeCV
import numpy as np

ridge_model = RidgeCV(
    alphas=np.logspace(-3,3,100),
    cv=5
)

ridge_model.fit(X_train_scaled, y_train)

y_pred_ridge = ridge_model.predict(X_test_scaled)

rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge))
r2_ridge = r2_score(y_test, y_pred_ridge)

print("Best Ridge Alpha:", ridge_model.alpha_)
```

```python
from sklearn.linear_model import LassoCV

lasso_model = LassoCV(
    cv=5,
    max_iter=10000
)

lasso_model.fit(X_train_scaled, y_train)

y_pred_lasso = lasso_model.predict(X_test_scaled)

rmse_lasso = np.sqrt(mean_squared_error(y_test, y_pred_lasso))
r2_lasso = r2_score(y_test, y_pred_lasso)

print("Best LASSO Alpha:", lasso_model.alpha_)
print("Nonzero Coefs:", sum(lasso_model.coef_ != 0))
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)

X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)

y_pred_poly = poly_model.predict(X_test_poly)

mse_poly = mean_squared_error(y_test, y_pred_poly)
rmse_poly = mse_poly ** 0.5
r2_poly = r2_score(y_test, y_pred_poly)
```

```python
from sklearn.decomposition import PCA

best_rmse = 999999
best_comp = 0

for n in range(5,31):

    pca = PCA(n_components=n)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_test_pca = pca.transform(X_test_scaled)

    model = LinearRegression()
    model.fit(X_train_pca, y_train)

    pred = model.predict(X_test_pca)

    rmse = np.sqrt(mean_squared_error(y_test,pred))

    if rmse < best_rmse:
        best_rmse = rmse
        best_comp = n

rmse_pcr = best_rmse
print("Best PCR Components:", best_comp)
```

```python
from sklearn.cross_decomposition import PLSRegression

best_rmse = 999999
best_comp = 0

for n in range(1,21):

    pls = PLSRegression(n_components=n)
    pls.fit(X_train_scaled, y_train)

    pred = pls.predict(X_test_scaled)

    rmse = np.sqrt(mean_squared_error(y_test,pred))

    if rmse < best_rmse:
        best_rmse = rmse
        best_comp = n

rmse_pls = best_rmse
print("Best PLS Components:", best_comp)
```

## Model Performance Comparison

Across all models tested, the standard linear regression model performed best, achieving the lowest RMSE of approximately 21,349 and the highest predictive accuracy on the test set. This suggests that the Ames Housing dataset contains a strong linear relationship between predictors and sale price.

The Ridge regression model performed second best with an RMSE of approximately 22,600, indicating that coefficient shrinkage helped maintain strong predictive performance while reducing potential overfitting.

The PLS regression model also performed well with an RMSE of approximately 23,454, outperforming both LASSO and PCR. This suggests that reducing predictors into latent components while incorporating the response variable can be useful for housing price prediction.

The LASSO regression model had an RMSE of approximately 23,602. Although it did not outperform linear regression, it remained competitive while shrinking many coefficients to zero, making it useful for variable selection and simpler interpretation.

## Effect of Regularization

Both Ridge and LASSO regression were used to penalize large coefficients and reduce overfitting.

Ridge kept all predictors while shrinking coefficient size.
LASSO selected a smaller subset of variables by forcing some coefficients to zero.

In this dataset, regularization did not outperform the baseline linear model, suggesting that the original predictors already captured the relationship effectively.

## Dimension Reduction Results

Principal Components Regression (PCR) achieved an RMSE of approximately 29,828, while Partial Least Squares (PLS) performed much better with an RMSE of 23,454.

This indicates that PLS was more effective because it forms components using both predictors and the response variable, whereas PCR only uses predictor variation.

## Nonlinear Model Performance

The GAM model achieved an RMSE of approximately 30,247, showing moderate predictive ability but not outperforming simpler linear methods.

The Polynomial regression model performed extremely poorly with an RMSE near 397,744, indicating severe overfitting caused by excessive model complexity.

These results suggest that nonlinear methods were unnecessary for this dataset.

## Cross-Validation Results

The 5-fold cross-validation RMSE for the linear model was approximately 35,252, which was higher than the single train-test split RMSE.

This suggests some variation across folds, but overall the linear model still generalized reasonably well.

AIC / BIC Results

AIC and BIC were calculated for the linear regression model:

AIC = 53,763
BIC = 55,324

These metrics help compare model fit while penalizing model complexity. Lower values indicate a better balance of fit and simplicity.

## Overall Conclusion

Overall, the results show that simple linear regression was the best predictive model for the Ames Housing dataset.

More complex methods such as GAM, PCR, and polynomial regression did not improve performance. Ridge, LASSO, and PLS remained competitive but did not exceed the baseline model.

This suggests that the dataset has a strong linear structure, making traditional regression methods highly effective for sale price prediction.
